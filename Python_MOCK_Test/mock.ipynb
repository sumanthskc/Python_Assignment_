```python
# reading dataframe 
import pandas as pd

data_frame=pd.read_csv("HousingDataset.csv")

data_frame.head(5)
```

```python
# Identify and remove the following columns: 
# a. ad_type 
# b. title 
# c. description 
# d. l4 
# e. l5 
# f. l6 

my_columns=data_frame.columns
data_frame.drop(columns=["ad_type","title","description","l4","l5","l6"],inplace=True)

after_delete=set(my_columns) - set(data_frame.columns)

```

```python
print(after_delete, "Deleted Successfully")
```

```python
# Investigate and discard all rows that contain Null values in any of these fields: 
# a. lon   
# b. lat   
# c. price_period   
# d. bedrooms   
# e. surface_total   
# f. rooms   
# g. price   
# h. surface_covered  

data_frame.dropna(subset=["lon","lat","price_period","bedrooms","surface_total","rooms","surface_covered"],inplace=True)
```

```python
Property_Details=pd.DataFrame(data_frame,columns=["id", "start_date", "end_date", "created_on", "lat", "lon", "l1", "l2", "l3", "rooms", "bedrooms", "bathrooms", 
"surface_total", "surface_covered"],index=None)
print(Property_Details.head(5))
```

```python
Property_Price_Details=pd.DataFrame(data_frame,columns=["id", "price", "currency", "price_period", "property_type", "operation_type"])
```

```python
Property_Price_Details.info()
```

```python
import sqlite3;
connection=sqlite3.connect("sql_assignment.db")
cursor=connection.cursor()
```

```python
Property_Details.to_sql("Property_details",connection,if_exists="replace",index=None)
Property_Price_Details.to_sql("Property_price_details",connection,if_exists="replace",index=None)
```

```python
# Retrieve properties that have a price greater than 1 million and are located in "Estados Unidos" (l1). 

name="Estados Unidos"

query1=""" SELECT * FROM 
        Property_Details p1 
        INNER JOIN 
        Property_Price_Details p2
        ON p1.id=p2.id 
        where p2.price>1000000
        and l1=?;
                """
cursor.execute(query1,(name,))
cursor.fetchmany()
for ele in cursor:
        print(ele)
```

```python
# Categorize properties based on their surface area as 'Small' if it's less than 50 square meters, 
# 'Medium' if it's 
# between 50 and 100 square meters, and 'Large' if it's greater than 100 square meters: 

query2="""Select id,surface_total,
          case 
                when surface_total<50 then "Small"
                when surface_total between 50 and 100 then "Medium"
                when surface_total>100 then "Large" 
          end
          as Category
          from Property_Details; """
cursor.execute(query2)
cursor.fetchmany()
for ele in cursor:
        print(ele)
```

```python
# List all properties (id) in the "Belgrano" neighborhood (l3) 
# that have the same number of bedrooms and 
# bathrooms as another property in the dataset:

query3="""SELECT Distinct d1.id as another_property FROM
          Property_Details d1
          JOIN
          Property_Details d2
          on 
          d1.id != d2.id
          and 
          d1.bedrooms=d2.bedrooms
          and 
          d1.bathrooms=d2.bathrooms
          where
          d1.l3="Belgrano";
"""
cursor.execute(query3)
cursor.fetchmany()
for ele in cursor:
    print(ele)
```

```python
# Calculate the average price per square meter (price / surface_total) for each property type (property_type) in 
# the "Belgrano" neighborhood (l3): 

query4="""SELECT p.property_type,
        AVG(p.price/ d.surface_total) as average_price
        FROM 
        Property_details d
        INNER JOIN
        Property_price_details p
        ON
        d.id=p.id
        WHERE d.l3="Belgrano"
        GROUP BY p.property_type;

"""
cursor.execute(query4)
cursor.fetchmany()
for ele in cursor:
    print(ele)
```

```python
# Identify properties that have a higher price than the average price of properties with the same number of 
# bedrooms and bathrooms. 

query5="""SELECT p.id,p.price 
            FROM Property_details d
            INNER JOIN
            Property_price_details p
            ON d.id = p.id
            and 
            p.price>( select avg(p1.price) as average 
            from Property_price_details p1 JOIN Property_Details d2
            on d2.id=p1.id
            where d2.bedrooms=d.bedrooms
            and d2.bathrooms=d.bathrooms)
""" 

cursor.execute(query5)
cursor.fetchmany()
for ele in cursor:
    print(ele)



```
